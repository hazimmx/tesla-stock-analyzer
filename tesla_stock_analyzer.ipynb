```python
!pip install yfinance #via yahoo finance
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ticker = yf.Ticker("TSLA")
info = ticker.info #the big dictionary data about Tesla
#print(info)

print("--- Company Info ---")
print("Company:", info['shortName'])
print("EPS:", info['epsTrailingTwelveMonths'])
print("PER:", info['trailingPE'])
print("Current Price:", info['currentPrice'])
print("Market Cap:", info['marketCap'])
print("52 Week High:", info['fiftyTwoWeekHigh'])
print("52 Week Low:", info['fiftyTwoWeekLow'])
```

```python
#EPS: earning per share
#PER: price to earning ratio -> investor paying [this value] wht tesla earns per year per share

eps = info['epsTrailingTwelveMonths']
pe = info['trailingPE']
fair_value = eps * pe

print(f"Estimated Fair Value for Tesla: ${fair_value:.2f}")
```

```python
hist = ticker.history(period="1y")
hist['Close'].plot(title="Tesla 1-Year Closing Price", figsize=(10,5))
plt.ylabel("Price ($)")
plt.grid(True)
plt.show()
```
